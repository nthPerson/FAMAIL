# latest_traffic.pkl Data Explorer

```python
import pickle
```

# Top level dictionary:
A list of keys representing a single state from the complete FAMAIL data model state-space: `(x_grid, y_grid, time_of_day, day_of_week)`
### Key index definition
- 0: x_grid index
- 1: y_grid index
- 2: time bucket $\in$ [0,287]
- 3: day-of-week index

```python
latest_traffic = pickle.load(open('../datasets/latest_traffic.pkl', 'rb'))

traffic_keys = list(latest_traffic.keys())
print(f'Number of keys: {len(traffic_keys)}')

for key_index in range(10):
    print(traffic_keys[key_index])
```

# Each key contains:
The traffic speed, waiting time, and one unverified value for the state, aggregated from historical data.
### Value index definition
- Key[0]: traffic speed
- Key[1]: waiting time
- Key[2]: unverified value 

```python
val_element_list = ['Traffic speed', 'Waiting time', '(?)unverified']

print(f"For example key {traffic_keys[0]}:")
for i, value in enumerate(latest_traffic[traffic_keys[0]]):
    print(f'Index [{i}] = {latest_traffic[traffic_keys[0]][i]} -- {val_element_list[i]}')
```
