# cgail_trajs2.pkl Data Exploration

```python
import pickle
```

## Top-level contains 50 driver objects

```python
cgail_trajs2 =  pickle.load(open('./Data/trajectory_info/cgail_trajs2.pkl', 'rb'))

driver_keys = list(cgail_trajs2.keys())
print(f'Number of drivers: {len(driver_keys)}')
print(cgail_trajs2.keys())
```

```python
for i, key in enumerate(driver_keys):
    print(f'Driver {i}: {key}')
```

## Each driver object contains a list of trajectories

```python
driver_0_trajs = cgail_trajs2[driver_keys[0]]

for i, traj in enumerate(driver_0_trajs):
    print(f"Traj_{i}: Length = {len(traj)}; states = {traj}")
```

## Each trajectory is a series of individual states

```python
driver_0_traj_0 =  cgail_trajs2[driver_keys[0]][0]

for i, state in enumerate(driver_0_traj_0):
    print(f'State {i}: {driver_0_traj_0[i]}')
```

## Each state is comprised of driver_id, x_grid, y_grid, time-of-day, day-of-week

```python
driver_0_traj_0_state_0 = cgail_trajs2[driver_keys[0]][0][0]

element_list = ['driver_id', 'x_grid', 'y_grid', 'time_bucket', 'day_index']

for i, element_name, element in zip(range(5), element_list, driver_0_traj_0_state_0):
    print(f'Index {i}: {element_name} = {element}')
```

## Total number of trajectories in cgail_trajs2

```python
len(cgail_trajs2[driver_keys[0]])

total_traj = 0
for key in driver_keys:
    total_traj += len(cgail_trajs2[key])

print(total_traj)
```
